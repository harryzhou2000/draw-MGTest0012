```python
from LogErrLoader import LoadDNDSLogErr
import numpy as np
import matplotlib.pyplot as plt


data_AOA5 = []
data_AOA5.append((LoadDNDSLogErr("NACA0012-AOA5_.log"), "GMRES-Jacobi"))
data_AOA5.append((LoadDNDSLogErr("NACA0012-AOA5-MGLP1_.log"), "GMRES-Jacobi-MGLP1"))
data_AOA5.append((LoadDNDSLogErr("NACA0012-AOA5-ILU2_.log"), "ILU2"))
data_AOA5.append((LoadDNDSLogErr("NACA0012-AOA5-ILU2-MGLP1_.log"), "ILU2-MGLP1"))
data_AOA5.append((LoadDNDSLogErr("NACA0012-AOA5-GILU2_.log"), "GMRES-ILU2"))

data_AOA15 = []
data_AOA15.append((LoadDNDSLogErr("NACA0012-AOA15_.log"), "GMRES-Jacobi"))
data_AOA15.append((LoadDNDSLogErr("NACA0012-AOA15-MGLP1_.log"), "GMRES-Jacobi-MGLP1"))
data_AOA15.append((LoadDNDSLogErr("NACA0012-AOA15-ILU2_.log"), "ILU2"))
data_AOA15.append((LoadDNDSLogErr("NACA0012-AOA15-ILU2-MGLP1_.log"), "ILU2-MGLP1"))
data_AOA15.append((LoadDNDSLogErr("NACA0012-AOA15-GILU2_.log"), "GMRES-ILU2"))

lwc = 1


def plotOne(data, label, xAxis="tWall", yAxis="res0"):
    plt.plot(data[xAxis][0:-1], data[yAxis][0:-1], label=label, lw=lwc)


def plotAfter(xlim, xlabel="t", ylabel="res0"):
    plt.yscale("log")
    leg = plt.legend()
    plt.xlim(xlim)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python

plt.figure(1001, figsize=(12, 8), dpi=160)
plt.title("AOA 5, res-time")
for (data, name) in data_AOA5:
    plotOne(data, name)
plotAfter((0,200))
```

```python
plt.figure(1002, figsize=(12, 8), dpi=160)
plt.title("AOA 5, res-niter")
for data, name in data_AOA5:
    plotOne(data, name, "iter")
plotAfter((0, 2000), xlabel="iter")
```
