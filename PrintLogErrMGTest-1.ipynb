```python
from LogErrLoader import LoadDNDSLogErr
import numpy as np
import matplotlib.pyplot as plt


data_AOA5 = []
data_AOA5.append(
    (LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA5.dir/ilu2_.log"), "ILU2")
)
data_AOA5.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA5.dir/gmres5x1-ilu2_.log"),
        "GMRES5x1-ILU2",
    )
)
data_AOA5.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA5.dir/ilu2-3ilu2_.log"),
        "ILU2-3[ILU2]",
    )
)
data_AOA5.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA5.dir/ilu2-3ilu2-3ilu2_.log"),
        "ILU2-3[ILU2]-3[ILU2]",
    )
)
data_AOA5.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA5.dir/ilu2-4ilu2_.log"),
        "ILU2-4[ILU2]",
    )
)
data_AOA5.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA5.dir/ilu2-3ilu2-3ilu1_.log"),
        "ILU2-3[ILU2]-3[LUSGS]",
    )
)

data_AOA15 = []
data_AOA15.append((LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA15.dir/ilu2_.log"), "ILU2"))
data_AOA15.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA15.dir/gmres5x1-ilu2_.log"),
        "GMRES5x1-ILU2",
    )
)
data_AOA15.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA15.dir/ilu2-3ilu2_.log"),
        "ILU2-3[ILU2]",
    )
)
data_AOA15.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA15.dir/ilu2-3ilu2-3ilu2_.log"),
        "ILU2-3[ILU2]-3[ILU2]",
    )
)
data_AOA15.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA15.dir/ilu2-4ilu2_.log"),
        "ILU2-4[ILU2]",
    )
)
data_AOA15.append(
    (
        LoadDNDSLogErr("NACA0012_H2-MGtest_1_AOA15.dir/ilu2-3ilu2-3ilu1_.log"),
        "ILU2-3[ILU2]-3[LUSGS]",
    )
)

lwc = 1

markEvery = 10

markerList = (
    ".",
    "o",
    "v",
    "^",
    "<",
    ">",
    "8",
    "s",
    "p",
    "*",
    "h",
    "H",
    "D",
    "d",
    "P",
    "X",
)

def plotOne(data, label, plotIndex=0, xAxis="tWall", yAxis="res0"):
    plt.plot(
        data[xAxis][0:-1],
        data[yAxis][0:-1],
        label=label,
        lw=lwc,
        marker=markerList[plotIndex],
        markevery=markEvery,
    )
    return plotIndex + 1


def plotAfter(xlim, xlabel="t", ylabel="res0"):
    plt.yscale("log")
    leg = plt.legend()
    plt.xlim(xlim)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
plt.figure(1001, figsize=(12, 8), dpi=160)
plt.title("AOA 5, res-time")
for i, (data, name) in enumerate(data_AOA5):
    plotOne(data, name, i)
plotAfter((0,100))
```

```python
plt.figure(1002, figsize=(12, 8), dpi=160)
plt.title("AOA 5, res-niter")
for i, (data, name) in enumerate(data_AOA5):
    plotOne(data, name, i, "iter")
plotAfter((0, 2000), xlabel="iter")
```
